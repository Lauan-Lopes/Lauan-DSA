```python
# Importando a biblioteca pandas

import pandas as pd
```

```python
# Importando dados csv

# file_name = 'Sales2022.csv'

caminho_arquivo = r'C:\Users\Administrator\CursoPython\Sales2022.csv'


# data_type converte colunas especificadas para o tipo de valor, str = string (cadeia de caracteres), coluna de data nao funciona no data_type

data_type = {'CustomerNumber': str,'SalesOrder': str}


# para ja converter quando eu for trazer o dataframe eu coloco: dtype = data_type, e tambem coloco: parse_dates = ['NomedaColuna'] para convertar colunas de data

df = pd.read_csv(caminho_arquivo, dtype = data_type, parse_dates = ['SalesDate'])
```

```python
# Cabecalho mostrando por padrao as 5 primeiras linhas do dataframe, mas colocando um numero dentro do parenteses apos o head(...) eu escolho quantas linhas quero exibir

df.head()
```

```python
# Funcao que mostra o tipo de dados para cada coluna: Int = Inteiro / Object = String (Cadeia de Caracteres) / Float = Flutuantes (Decimais)

df.dtypes
```

```python
# Exibe as colunas do meu dataframe

df.columns

# Seleciona as colunas atraves de uma variavel chamada "Colunas"

colunas = ['Country', 'State', 'City',
       'PlantName', 'MaterialNumber', 'BusinessLine', 'Category', 'SalesDate',
       'SalesOrder', 'Volume', 'Revenue']

# Exibe as colunas selecionadas

df[colunas]

# Das colunas selecionadas eu seleciono mais ainda apenas as que eu quero com o [[]]

df[['MaterialNumber','Category' ]].head()
```

Filtrando dados

```python
# Unique pega os valores distintos da coluna do meu dataframe

# df['PlantName'].unique()

# Filtro apenas os valores que a PlantName e igual a Sorocaba e armazeno na variavel sorocaba_df, para isso devo escrever o nome da variavel e escrever "df[df['']=='']]"

sorocaba_df = df[df['PlantName']=='Sorocaba']
sorocaba_df
```

```python
# Pega o dataframe de sorocaba_df e filtra apenas os que o Revenue e maior ou igual a 1000 usando a sintaxe "variavel[variavel['coluna']>=valor]"

sorocaba_df[sorocaba_df['Revenue']>=1000].head()
```

```python
# Pega o dataframe de sorocaba_df e filtra apenas os que a data que e maior ou igual a 2022-01-01 usando a sintaxe "variavel[variavel['coluna']>='valor']"

sorocaba_df[sorocaba_df['SalesDate']>='2022-01-01']
```

```python
# Pega o dataframe de sorocaba_df e filtra apenas os que a data que e maior ou igual a 2022-01-01 e menor ou igual 2022-01-31
# Usando a sintaxe "variavel[(variavel['coluna']>='valor') & variavel[(variavel['coluna']<='valor')]" entre parenteses sao os criterios que eu adotei para o filtro

sorocaba_df[(sorocaba_df['SalesDate']>='2022-01-01') & (sorocaba_df['SalesDate']<= '2022-01-31')]
```

```python
# Pega o dataframe de sorocaba_df da modificacao acima e classifica em ordem decrescente com a funcao ".sort_values(by='coluna', ascending=False)" 
# Se eu colocar .sort_values(by='coluna', ascending=True) vai filtrar em ordem crescente a coluna especificada no "(by='coluna')"

sorocaba_df[(sorocaba_df['SalesDate']>='2022-01-01') & (sorocaba_df['SalesDate']<= '2022-01-31')].sort_values(by='SalesDate', ascending=False)
```

Group by

```python
# Agrupo os valores pela categoria somando a Revenue com a sintaxe "variavel.groupby('coluna')['coluna'].sum()"
# A primeira coluna nos parenteses e a coluna que quero ver agrupado pela segunda no conchetes e no final a operacao de agrupamento .sum() 

sorocaba_df.groupby('Category')['Revenue'].sum()
```

```python
# Coloco ".sort_values(ascending=False)" para colocar em ordem do maior para o menor

sorocaba_df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
# Armazeno em uma variavel a funcao de group by que agrupa por 2 colunas na parte de ".groupby(['BusinessLine', 'Category'])", onde [] e a lista de colunas agrupadas

sorocaba_df_group = sorocaba_df.groupby(['BusinessLine', 'Category'])['Revenue'].sum().sort_values(ascending=False)

# Type eu visualizo o tipo de dados armazenados na variavel "sorocaba_df_group"

type(sorocaba_df_group)
```

```python
# Transforma o Series do pandas para um dataframe com o final no codigo ".reset_index()"

sorocaba_df_group = sorocaba_df.groupby(['BusinessLine', 'Category'])['Revenue'].sum().sort_values(ascending=False).reset_index()
sorocaba_df_group
```

```python
# Exibe validando o tipo da variavel agora que transformei em dataframe

type(sorocaba_df_group)
```

Tabelas Dinamicas

```python
# Crio uma tabela dinamica do meu dataframe df que me mostra o total de aditivo e lubrificante vendidos por estado
# Na sintaxe eu coloco "df.pixot_table" no "index" vai ser as minhas linhas no "colunms" minhas colunas "value" os valores 
# No "aggfunc" e a operacao de agrupamento se vai ser soma media ou outro no "margins = true" sao os totais por linha e coluna e "margins_name" o nome dessa linha e coluna nova

df.pivot_table(
    index = 'State',
    columns = 'BusinessLine',
    values = 'Revenue',
    aggfunc = 'sum',
    margins = True, 
    margins_name = 'Total' 
)
```
